```python
import numpy as np
import matplotlib.pyplot as plt

def funcao_quadratica(a, b, c):
    """Calcula vértice, raízes e intercepto de uma função quadrática"""

    # Vértice
    xv = -b / (2 * a)
    yv = a * xv**2 + b * xv + c
    vertice = (xv, yv)

    # Raízes (usando Bhaskara)
    delta = b**2 - 4*a*c
    if delta >= 0:
        x1 = (-b + np.sqrt(delta)) / (2 * a)
        x2 = (-b - np.sqrt(delta)) / (2 * a)
        raizes = (x1, x2)
    else:
        raizes = "Sem raízes reais"

    # Intercepto Y
    intercepto_y = c

    return vertice, raizes, intercepto_y

# Exemplo: f(x) = x² - 4x + 3
a, b, c = 1, -4, 3
vertice, raizes, intercepto_y = funcao_quadratica(a, b, c)

print(f"Vértice: {vertice}")
print(f"Raízes: {raizes}")
print(f"Intercepto Y: {intercepto_y}")
```

```python
def plotar_parabola(a, b, c):
    """Plota a função quadrática"""

    # Criando um intervalo de x
    x = np.linspace(-10, 10, 400)
    y = a * x**2 + b * x + c

    # Plot
    plt.figure(figsize=(8,6))
    plt.plot(x, y, label=f'$f(x) = {a}x^2 + {b}x + {c}$', color='blue')

    # Marcar vértice
    vertice, raizes, intercepto_y = funcao_quadratica(a, b, c)
    plt.scatter(*vertice, color='red', label=f'Vértice {vertice}')

    # Marcar raízes (se existirem)
    if isinstance(raizes, tuple):
        plt.scatter(raizes[0], 0, color='green', label=f'Raiz {raizes[0]:.2f}')
        plt.scatter(raizes[1], 0, color='green', label=f'Raiz {raizes[1]:.2f}')

    # Marcar intercepto Y
    plt.scatter(0, intercepto_y, color='orange', label=f'Intercepto Y ({intercepto_y})')

    # Configurações do gráfico
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.legend()
    plt.title("Gráfico da Função Quadrática")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()

# Exemplo
plotar_parabola(a, b, c)
```
